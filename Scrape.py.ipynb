```python
import pandas as pd
import numpy as np
from random import randint
from time import sleep
```

```python
sp500 = pd.read_html('https://en.m.wikipedia.org/wiki/List_of_S%26P_500_companies')
```

```python
sp500 = sp500[0]
```

```python
sp500 = sp500.loc[:, [0,1,3,4]]
colnames = list(sp500.loc[0,:])  #col names are in first row
```

```python
sp500.set_axis(colnames, axis='columns', inplace=True)
sp500 = sp500.iloc[1:]
```

```python
sp500.to_csv('/Users/michaelsankari/Documents/NYC Data Science/Earnings Shiny/data/spy.csv', index=False)
```

```python
tickers = list(sp500['Symbol'])
```

```python
url_list = ['https://www.optionslam.com/earnings/stocks/' + ticker for ticker in tickers]     
```

```python
i=0
for ticker, url in zip(tickers, url_list):
    print(str(i) + ' ' + url + ' ' + ticker) 
    i += 1
```

```python
for ticker, url in zip(tickers[85:], url_list[85:]):
    print('starting ' + url + ' ' + ticker)
    df = pd.read_html(url)
    df = df[4]
    df = df[0]
    df.dropna(inplace=True)
    df = df[np.logical_or(df.str.contains('AC'), df.str.contains('BO'))]
    df.to_csv('/Users/michaelsankari/Documents/NYC Data Science/Earnings Shiny/data/dates/'+ticker+'.csv', 
              header=False, index=False)
    print('downloaded ' + str(len(df.index)) + ' dates')
    sleep(randint(2,5)) #don't overload server
    
```
